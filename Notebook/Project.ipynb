```python
import pandas as pd 

df= pd.read_excel("ParalympicsData1.xlsx")

df.head(50)
```

```python
years= df["Year"].unique()

for listYears in years:
    print(listYears)
```

```python
ranking= df[["Total", "Country"]].groupby("Country").sum("Total").sort_values(by="Total", ascending=False)
ranking.head(5)
```

```python
lastEditions= df[(df["Year"]>=2000) & (df["Year"]<= 2018)]
bones= lastEditions. groupby("Country")["Total"].sum().reset_index()
grave= bones.sort_values(by="Total", ascending=False)
road= grave.head(10)
print(road.to_string(index=False))
```

```python
ranking= df[["Total", "Country"]].groupby("Country").sum("Total").sort_values(by="Total", ascending=True)
ranking.head(10)
```
